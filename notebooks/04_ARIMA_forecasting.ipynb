```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, r2_score

# üìå 1Ô∏è‚É£ ÌÖåÏä¨Îùº(TSLA) Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥Î°úÎìú (ÏµúÍ∑º 2ÎÖÑÏπò)
df = yf.download("TSLA", start="2022-01-01", end="2024-01-01")
df = df.reset_index()

# üìå 2Ô∏è‚É£ ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï† (80% ÌïôÏäµ, 20% ÌÖåÏä§Ìä∏)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size].copy(), df.iloc[train_size:].copy()

# Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏ (ÎîîÎ≤ÑÍπÖÏö©)
print(f"Train DataFrame: {train.shape}")
print(f"Test DataFrame: {test.shape}")
print(train.head())

# üìå 3Ô∏è‚É£ ÏµúÏ†ÅÏùò (p, d, q) Í∞í Ï∞æÍ∏∞
auto_model = auto_arima(train["Close"], seasonal=False, stepwise=True, trace=True)
best_p, best_d, best_q = auto_model.order
print(f"‚úÖ ÏµúÏ†ÅÏùò (p, d, q) Í∞í: (p={best_p}, d={best_d}, q={best_q})")

# üìå 4Ô∏è‚É£ ARIMA Î™®Îç∏ ÌïôÏäµ
model = ARIMA(train["Close"], order=(best_p, best_d, best_q))
model_fit = model.fit()

# üìå 5Ô∏è‚É£ ÏòàÏ∏° ÏàòÌñâ
train_pred = model_fit.predict(start=best_p, end=len(train) - 1, typ="levels")
test_pred = model_fit.forecast(steps=len(test))  # Î°§ÎßÅ ÏòàÏ∏°Ïù¥ ÏïÑÎãàÎùº, Ï†ÑÏ≤¥ ÏòàÏ∏°

# üìå 6Ô∏è‚É£ ÏÑ±Îä• ÌèâÍ∞Ä
train_actual = train["Close"][best_p:]
test_actual = test["Close"]

train_mse = mean_squared_error(train_actual, train_pred)
test_mse = mean_squared_error(test_actual, test_pred)
train_r2 = r2_score(train_actual, train_pred)
test_r2 = r2_score(test_actual, test_pred)

print(f"üìå Training MSE: {train_mse:.4f}, R¬≤: {train_r2:.4f}")
print(f"üìå Test MSE: {test_mse:.4f}, R¬≤: {test_r2:.4f}")

# üìå 7Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏°)
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Actual", color="black")
plt.plot(test["Date"], test_pred, label="Predicted (ARIMA Model)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA Model")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå 8Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í)
plt.figure(figsize=(12, 5))
plt.plot(train["Date"][best_p:], train_actual, label="Actual (Training)", color="black")
plt.plot(train["Date"][best_p:], train_pred, label="Predicted (Training)", color="blue", linestyle="--")
plt.title("Tesla Stock Price Prediction using ARIMA - Training Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå 9Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í)
plt.figure(figsize=(12, 5))
plt.plot(test["Date"], test_actual, label="Actual (Test)", color="black")
plt.plot(test["Date"], test_pred, label="Predicted (Test)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA - Test Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# ‚úÖ ÏµúÏ†Å (p, d, q) ÌÉêÏÉâ + ARIMA Î™®Îç∏ ÌïôÏäµ + Î°§ÎßÅ ÏòàÏ∏° ÏΩîÎìú

‚úÖ Í∞úÏÑ†Îêú Í∏∞Îä•
ÏµúÏ†ÅÏùò (p, d, q) Í∞í ÏûêÎèô ÌÉêÏÉâ
pmdarima.auto_arima()Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÏµúÏ†ÅÏùò ÌååÎùºÎØ∏ÌÑ∞Î•º ÏûêÎèôÏúºÎ°ú Ï∞æÏùå.
ARIMA Î™®Îç∏ ÌïôÏäµ Î∞è Ï†ÑÏ≤¥ ÏòàÏ∏°
model_fit.predict()Î°ú ÌõàÎ†® Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° ÏàòÌñâ.
Î°§ÎßÅ ÏòàÏ∏° (One-step rolling forecast)
ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Ìïú Ïä§ÌÖùÏî© ÏòàÏ∏°ÌïòÎäî Î∞©Ïãù Ï†ÅÏö©.
ÏÑ±Îä• ÌèâÍ∞Ä Î∞è ÏãúÍ∞ÅÌôî
MSE(Mean Squared Error) Î∞è R¬≤(ÏÑ§Î™ÖÎ†•) Í≥ÑÏÇ∞
ÌõàÎ†® Îç∞Ïù¥ÌÑ∞, ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏòàÏ∏° Í≤∞Í≥º ÏãúÍ∞ÅÌôî.

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, r2_score

# üìå 1Ô∏è‚É£ ÌÖåÏä¨Îùº(TSLA) Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥Î°úÎìú (ÏµúÍ∑º 2ÎÖÑÏπò)
df = yf.download("TSLA", start="2022-01-01", end="2024-01-01")

# ‚úÖ Îã§Ï§ë Ïù∏Îç±Ïä§ Ï†úÍ±∞ (ÌïÑÏàò)
df.columns = df.columns.droplevel(1)  # MultiIndex Ï†úÍ±∞
df = df.reset_index()  # Ïù∏Îç±Ïä§ Ï¥àÍ∏∞Ìôî

# üìå ‚úÖ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ïª¨Îüº ÌôïÏù∏ (ÎîîÎ≤ÑÍπÖ)
print("üîç DataFrame Columns:", df.columns)
print(df.head())

# üìå 2Ô∏è‚É£ ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï† (80% ÌïôÏäµ, 20% ÌÖåÏä§Ìä∏)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size].copy(), df.iloc[train_size:].copy()

# üìå 3Ô∏è‚É£ ÏµúÏ†ÅÏùò (p, d, q) Í∞í Ï∞æÍ∏∞
auto_model = auto_arima(train["Close"], seasonal=False, stepwise=True, trace=True)
best_p, best_d, best_q = auto_model.order
print(f"‚úÖ ÏµúÏ†ÅÏùò (p, d, q) Í∞í: (p={best_p}, d={best_d}, q={best_q})")

# üìå 4Ô∏è‚É£ ARIMA Î™®Îç∏ ÌïôÏäµ (Ï†ÑÏ≤¥ ÏòàÏ∏°Ïö©)
model = ARIMA(train["Close"].astype(float), order=(best_p, best_d, best_q))  # ‚úÖ float Î≥ÄÌôò
model_fit = model.fit()

# üìå 5Ô∏è‚É£ ÏòàÏ∏° ÏàòÌñâ
train_pred = model_fit.predict(start=best_p, end=len(train) - 1, typ="levels")
test_pred = model_fit.forecast(steps=len(test))

# üìå 6Ô∏è‚É£ ÏÑ±Îä• ÌèâÍ∞Ä
train_actual = train["Close"].iloc[best_p:]  # Ïã§Ï†úÍ∞í
test_actual = test["Close"]

train_mse = mean_squared_error(train_actual, train_pred)
test_mse = mean_squared_error(test_actual, test_pred)
train_r2 = r2_score(train_actual, train_pred)
test_r2 = r2_score(test_actual, test_pred)

print(f"üìå Training MSE: {train_mse:.4f}, R¬≤: {train_r2:.4f}")
print(f"üìå Test MSE: {test_mse:.4f}, R¬≤: {test_r2:.4f}")

# üìå 7Ô∏è‚É£ Î°§ÎßÅ ÏòàÏ∏° (One-step rolling forecast)
rolling_predictions = []
history = list(train["Close"].dropna().astype(float))  # ‚úÖ float Î≥ÄÌôò Î∞è NaN Ï†úÍ±∞

# üìå ‚úÖ history Î¶¨Ïä§Ìä∏ ÌôïÏù∏ (ÎîîÎ≤ÑÍπÖ)
print("üîç Initial History Values:", history[:10])

for t in range(len(test)):
    history_array = np.array(history, dtype=float)  # ‚úÖ NumPy Î∞∞Ïó¥ Î≥ÄÌôò
    model = ARIMA(history_array, order=(best_p, best_d, best_q))
    model_fit = model.fit()
    pred = model_fit.forecast(steps=1)[0]  # Ìïú Ïä§ÌÖù ÏòàÏ∏°
    rolling_predictions.append(pred)

    # ‚úÖ `float()` Î≥ÄÌôò ÌõÑ Ï∂îÍ∞Ä (ÎîîÎ≤ÑÍπÖ)
    try:
        history.append(float(test["Close"].iloc[t]))  # float Î≥ÄÌôòÌïòÏó¨ Ï∂îÍ∞Ä
    except ValueError:
        print(f"‚ùå Î≥ÄÌôò Ïã§Ìå®: test['Close'].iloc[{t}] = {test['Close'].iloc[t]}")  # ÏóêÎü¨ Î©îÏãúÏßÄ Ï∂úÎ†•

# üìå 8Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏°)
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Actual", color="black")
plt.plot(test["Date"], rolling_predictions, label="Predicted (ARIMA Model)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA Model (Full Forecast)")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå 9Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í)
plt.figure(figsize=(12, 5))
plt.plot(train["Date"].iloc[best_p:], train_actual, label="Actual (Training)", color="black")
plt.plot(train["Date"].iloc[best_p:], train_pred, label="Predicted (Training)", color="blue", linestyle="--")
plt.title("Tesla Stock Price Prediction using ARIMA - Training Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå üîü Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í - Î°§ÎßÅ ÏòàÏ∏°)
plt.figure(figsize=(12, 5))
plt.plot(test["Date"], test_actual, label="Actual (Test)", color="black")
plt.plot(test["Date"], rolling_predictions, label="Rolling Predicted (Test)", linestyle="--", color="green")
plt.title("Tesla Stock Price Prediction using ARIMA - Rolling Forecast")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

## Full series

Ï†ÑÏ≤¥ Íµ¨Í∞Ñ(ÌïôÏäµ+ÌÖåÏä§Ìä∏)ÏùÑ ÌïúÍ∫ºÎ≤àÏóê Ïç®ÏÑú ÌååÎùºÎØ∏ÌÑ∞ Ï∂îÏ†ï ‚Üí ÎØ∏Îûò HÏùº ÏòàÏ∏°.
ÌèâÍ∞ÄÏö©ÏúºÎ°ú Ïì∞Î©¥ Îç∞Ïù¥ÌÑ∞ ÎàÑÏàòÎùºÏÑú Î≥¥ÌÜµ ‚ÄúÏµúÏ¢Ö Ï†ÅÌï©/ÏãúÎÇòÎ¶¨Ïò§ ÏÉùÏÑ±‚ÄùÏóê ÏÇ¨Ïö©.

## Fixed-origin(Ï†ïÏ†Å ÌååÎùºÎØ∏ÌÑ∞)

ÌïôÏäµ Íµ¨Í∞ÑÏóêÏÑú (p,q)ÏôÄ Í≥ÑÏàòÎ•º Ìïú Î≤àÎßå Ï∂îÏ†ïÌïòÍ≥† ÌÖåÏä§Ìä∏ Ï†ÑÏ≤¥Ïóê Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©.
Í∞ÑÎã®¬∑Îπ†Î¶Ñ, ÌïòÏßÄÎßå Íµ≠Î©¥ Î≥ÄÌôîÏóî ÎëîÍ∞ê.

## Rolling window

Îß§ ÏãúÏ†êÎßàÎã§ ÏµúÍ∑º WÍ∞úÎßå ÏÇ¨Ïö©Ìï¥ 1-step ÏòàÏ∏°(Ï∞ΩÏù¥ ÏïûÏúºÎ°ú Î∞ÄÎ¶º).
ÏµúÏã† Î≥ÄÌôîÏóê ÎØºÍ∞ê(ÎìúÎ¶¨ÌîÑÌä∏ ÎåÄÏùë), Ïò§ÎûòÎêú Ï†ïÎ≥¥Îäî Î≤ÑÎ¶º.

## Expanding window

ÏãúÏûëÎ∂ÄÌÑ∞ ÌòÑÏû¨ ÏßÅÏ†ÑÍπåÏßÄ ÎàÑÏ†Å Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤¥Î°ú Îß§Î≤à Îã§Ïãú Ï∂îÏ†ï ÌõÑ 1-step ÏòàÏ∏°.
Îç∞Ïù¥ÌÑ∞ Ï∂ïÏ†Å ‚Üí ÏïàÏ†ïÏÑ±‚Üë(Î∂ÑÏÇ∞‚Üì), ÌïòÏßÄÎßå Î≥ÄÌôî Ï†ÅÏùëÏù¥ ÎäêÎ¶¥ Ïàò ÏûàÍ≥† Í≥ÑÏÇ∞Îüâ‚Üë.

```python
# ==================== Colab Ï§ÄÎπÑ ====================
%pip -q install yfinance statsmodels scikit-learn pandas numpy matplotlib

import warnings
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

from sklearn.metrics import r2_score
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tools.sm_exceptions import ConvergenceWarning

# Í≤ΩÍ≥† Ï†ïÎ¶¨
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=ConvergenceWarning)
warnings.filterwarnings("ignore", message="Non-invertible")
warnings.filterwarnings("ignore", message="Maximum Likelihood")
warnings.filterwarnings("ignore", message="No frequency information")
plt.rcParams["figure.figsize"] = (12, 5)

# ==================== Îç∞Ïù¥ÌÑ∞ Î°úÎìú ====================
symbol     = "TSLA"
start_date = "2022-01-01"
end_date   = "2025-09-30"

px = yf.download(symbol, start=start_date, end=end_date).reset_index()
px = px[["Date", "Close"]].dropna().copy()
px["Date"] = pd.to_datetime(px["Date"])

y = px["Close"].to_numpy(dtype=float)
dates = px["Date"].to_numpy()

# ÌïôÏäµ/ÌÖåÏä§Ìä∏ Î∂ÑÌï† (Í∞ÄÍ≤© Î†àÎ≤® Í∏∞Ï§Ä 80/20)
split_idx = int(len(y) * 0.8)
y_train, y_test = y[:split_idx], y[split_idx:]
dates_train, dates_test = dates[:split_idx], dates[split_idx:]

print(f"Samples: total={len(y)}, train={len(y_train)}, test={len(y_test)}")

# ==================== Ïú†Ìã∏Ìï®Ïàò ====================
def rmse(a, b):
    a = np.asarray(a, dtype=float); b = np.asarray(b, dtype=float)
    return float(np.sqrt(np.mean((a - b) ** 2)))

def mae(a, b):
    a = np.asarray(a, dtype=float); b = np.asarray(b, dtype=float)
    return float(np.mean(np.abs(a - b)))

def r2(a, b):
    return float(r2_score(np.asarray(a, dtype=float), np.asarray(b, dtype=float)))

def fit_arima_safe(y_win, order, maxiter=200):
    """ARIMA(p,d,q) ÏïàÏ†Ñ Ï†ÅÌï©. Ïã§Ìå® Ïãú None."""
    p, d, q = [int(x) for x in order]
    try:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            model = ARIMA(
                y_win, order=(p, d, q), trend="n",
                enforce_stationarity=False, enforce_invertibility=False
            )
            fit = model.fit(method_kwargs={"maxiter": maxiter, "disp": 0})
        return fit
    except Exception:
        return None

def select_arima_order(y_win, p_max=5, d_set=(0,1,2), q_max=5):
    """
    Í∞ÑÎã® AIC Í∑∏Î¶¨ÎìúÏÑúÏπòÎ°ú (p,d,q) ÏÑ†ÌÉù.
    ÏàòÎ†¥ Ïã§Ìå®/ÎπÑÏ†ïÏÉÅ ÏºÄÏù¥Ïä§Îäî Í±¥ÎÑàÎúÄ. ÌïòÎÇòÎèÑ ÏïàÎêòÎ©¥ (1,1,1) Î∞òÌôò.
    """
    best = None
    best_aic = np.inf
    y_win = np.asarray(y_win, dtype=float)
    for d in d_set:
        for p in range(p_max + 1):
            for q in range(q_max + 1):
                if (p, d, q) == (0, 0, 0):
                    continue
                fit = fit_arima_safe(y_win, (p, d, q), maxiter=300)
                if fit is None:
                    continue
                aic = getattr(fit, "aic", np.inf)
                if np.isfinite(aic) and aic < best_aic:
                    best_aic = aic
                    best = (p, d, q)
    return best if best is not None else (1, 1, 1)

# ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞
P_MAX, Q_MAX = 5, 5
D_SET = (0, 1, 2)
W = 180                 # Rolling Ï∞Ω Í∏∏Ïù¥(ÏòÅÏóÖÏùº ÏïΩ 9Í∞úÏõî Í∞ÄÏ†ï)
RESELECT_STRIDE = 10    # (p,d,q) Ïû¨ÏÑ†ÌÉù Í∞ÑÍ≤©(Í≥ÑÏÇ∞Îüâ Ï†úÏñ¥)
H_FUTURE = 20           # Full-series ÎØ∏Îûò ÏòàÏ∏° Í∏∏Ïù¥(Ïùº)

# ================================================================
# A) Full series ‚Äî Ï†ÑÏ≤¥Î°ú (p,d,q) ÏÑ†ÌÉù, Ï†ÑÏ≤¥Î°ú Ï†ÅÌï© ‚Üí ÎØ∏Îûò H_FUTURE ÏòàÏ∏°
#   * ÌèâÍ∞ÄÏö©ÏúºÎ°úÎäî Îç∞Ïù¥ÌÑ∞ ÎàÑÏàò ÏúÑÌóò! ÏµúÏ¢Ö ÏãúÎÇòÎ¶¨Ïò§ ÏÉùÏÑ±Ïö©ÏúºÎ°ú Î≥¥ÏÑ∏Ïöî.
# ================================================================
best_full = select_arima_order(y, p_max=P_MAX, d_set=D_SET, q_max=Q_MAX)
print(f"[Full] best (p,d,q) by AIC: {best_full}")

fit_full = fit_arima_safe(y, best_full, maxiter=400)
if fit_full is not None:
    y_future = fit_full.forecast(steps=H_FUTURE).astype(float)
else:
    # ÏôÑÏ†ÑÌûà Ïã§Ìå®ÌïòÎ©¥ ÎßàÏßÄÎßâ Í∞íÏúºÎ°ú ÌèâÌÉÑ Ïó∞Ïû•
    y_future = np.full(H_FUTURE, y[-1], dtype=float)

future_dates = pd.date_range(dates[-1] + pd.Timedelta(days=1), periods=H_FUTURE, freq="D")

plt.figure()
plt.plot(dates, y, label="Actual", linewidth=1.4, color="black")
plt.plot(future_dates, y_future, label=f"Full ARIMA{best_full}", linestyle="--")
plt.title(f"{symbol} ‚Äî Full-series ARIMA{best_full} (level) future {H_FUTURE}d")
plt.xlabel("Date"); plt.ylabel("Price"); plt.xticks(rotation=45); plt.legend(); plt.tight_layout(); plt.show()

print("\n[A] Full series ÏÑ§Î™Ö")
print("- Ï†ÑÏ≤¥ Íµ¨Í∞ÑÏùÑ Ìïú Î≤àÏóê ÏÇ¨Ïö©Ìï¥ (p,d,q)ÏôÄ Í≥ÑÏàòÎ•º Ï∂îÏ†ïÌïòÍ≥†, ÏïûÏúºÎ°ú H_FUTUREÏùºÏùÑ ÏòàÏ∏°Ìï©ÎãàÎã§.")
print("- ÌèâÍ∞ÄÏóê Ïì∞Î©¥ Îç∞Ïù¥ÌÑ∞ ÎàÑÏàòÏù¥ÎØÄÎ°ú ÏµúÏ¢Ö Ï†ÅÌï©/ÏãúÎÇòÎ¶¨Ïò§ ÏÉùÏÑ± Ïö©ÎèÑÎ°ú Í∂åÏû•.\n")

# ====================================================================
# B) Fixed-origin(Ï†ïÏ†Å ÌååÎùºÎØ∏ÌÑ∞) ‚Äî ÌïôÏäµÏóêÏÑú Ìïú Î≤àÎßå Ï∂îÏ†ï, ÌÖåÏä§Ìä∏ Ï†ÑÍµ¨Í∞Ñ Í∑∏ÎåÄÎ°ú ÏÇ¨Ïö©
# ====================================================================
best_fix = select_arima_order(y_train, p_max=P_MAX, d_set=D_SET, q_max=Q_MAX)
print(f"[Fixed-origin] best (p,d,q) by AIC: {best_fix}")

fit_static = fit_arima_safe(y_train, best_fix, maxiter=400)
if fit_static is not None:
    y_pred_static = fit_static.forecast(steps=len(y_test)).astype(float)
else:
    y_pred_static = np.full(len(y_test), y_train[-1], dtype=float)

rmse_b = rmse(y_test, y_pred_static)
mae_b  = mae(y_test, y_pred_static)
r2_b   = r2(y_test, y_pred_static)
print(f"[Fixed-origin] RMSE={rmse_b:.4f} | MAE={mae_b:.4f} | R2={r2_b:.4f}")

plt.figure()
plt.plot(dates_test, y_test, label="Actual (Test)", color="black", linewidth=1.4)
plt.plot(dates_test, y_pred_static, label=f"Fixed-origin ARIMA{best_fix}", linestyle="--")
plt.title(f"{symbol} ‚Äî Fixed-origin (static) ARIMA{best_fix}")
plt.xlabel("Date"); plt.ylabel("Price"); plt.xticks(rotation=45); plt.legend(); plt.tight_layout(); plt.show()

print("\n[B] Fixed-origin(Ï†ïÏ†Å ÌååÎùºÎØ∏ÌÑ∞) ÏÑ§Î™Ö")
print("- ÌïôÏäµ Íµ¨Í∞ÑÏóêÏÑúÎßå (p,d,q)ÏôÄ Í≥ÑÏàòÎ•º Ï∂îÏ†ïÌïòÍ≥†, ÌÖåÏä§Ìä∏ Íµ¨Í∞Ñ Ï†ÑÏ≤¥Ïóê ÏóÖÎç∞Ïù¥Ìä∏ ÏóÜÏù¥ ÏÇ¨Ïö©Ìï©ÎãàÎã§.")
print("- Îã®Ïàú/Îπ†Î¶Ñ, Í∑∏Îü¨ÎÇò Î†àÏßê Î≥ÄÎèôÏóê ÎëîÍ∞ê.\n")

# ====================================================================
# C) Rolling window ‚Äî ÏµúÍ∑º WÍ∞úÎ°ú Îß§ÏãúÏ†ê Ïû¨Ï†ÅÌï© (p,d,qÎäî Ï£ºÍ∏∞Ï†ÅÏúºÎ°ú Ïû¨ÏÑ†ÌÉù)
# ====================================================================
pred_roll = np.full(len(y_test), np.nan, dtype=float)
hist = y_train.copy()
current_order = None

for i in range(len(y_test)):
    y_win = hist[-W:] if len(hist) >= W else hist
    # Ï£ºÍ∏∞Ï†Å (p,d,q) Ïû¨ÏÑ†ÌÉù
    if (i % RESELECT_STRIDE == 0) or (current_order is None):
        current_order = select_arima_order(y_win, p_max=P_MAX, d_set=D_SET, q_max=Q_MAX)
    # 1-step ÏòàÏ∏°
    fit = fit_arima_safe(y_win, current_order, maxiter=300)
    if fit is not None:
        pred_roll[i] = float(fit.forecast(steps=1)[0])
    else:
        # Î∞±ÏóÖ: Îã®Ïàú ÌèâÍ∑†(Ï∞®Î∂Ñ Ï∞®ÏàòÏóê Î¨¥Í¥ÄÌïòÍ≤å Î†àÎ≤® ÌèâÍ∑† ÏÇ¨Ïö©)
        pred_roll[i] = float(np.mean(y_win[-min(10, len(y_win)):]))
    # ÏõåÌÅ¨-Ìè¨ÏõåÎìú(Ïã§Ï†úÍ∞í Ï∂îÍ∞Ä)
    hist = np.append(hist, y_test[i])

rmse_c = rmse(y_test, pred_roll)
mae_c  = mae(y_test, pred_roll)
r2_c   = r2(y_test, pred_roll)
print(f"[Rolling window] RMSE={rmse_c:.4f} | MAE={mae_c:.4f} | R2={r2_c:.4f}")

plt.figure()
plt.plot(dates_test, y_test, label="Actual (Test)", color="black", linewidth=1.4)
plt.plot(dates_test, pred_roll, label=f"Rolling ARIMA (W={W}, stride={RESELECT_STRIDE})", linestyle="--")
plt.title(f"{symbol} ‚Äî Rolling-window ARIMA (level)")
plt.xlabel("Date"); plt.ylabel("Price"); plt.xticks(rotation=45); plt.legend(); plt.tight_layout(); plt.show()

print("\n[C] Rolling window ÏÑ§Î™Ö")
print("- Í≥†Ï†ï Í∏∏Ïù¥ WÏùò ÏµúÍ∑º Îç∞Ïù¥ÌÑ∞Îßå ÏÇ¨Ïö©Ìï¥ Îß§ ÏãúÏ†ê 1-step ÏòàÏ∏°(Ï∞ΩÏù¥ ÏïûÏúºÎ°ú Ïù¥Îèô).")
print("- ÏµúÏã† Î≥ÄÌôîÏóê ÎØºÍ∞ê(ÎìúÎ¶¨ÌîÑÌä∏ ÎåÄÏùë), Ïò§ÎûòÎêú Ï†ïÎ≥¥Îäî Î≤ÑÎ†§ Îç∞Ïù¥ÌÑ∞ ÎÇ≠ÎπÑ Í∞ÄÎä•.\n")

# ====================================================================
# D) Expanding window ‚Äî ÏãúÏûë~ÌòÑÏû¨ ÏßÅÏ†ÑÍπåÏßÄ ÎàÑÏ†ÅÎç∞Ïù¥ÌÑ∞Î°ú 1-step ÏòàÏ∏°
#     (p,d,qÎäî Ï£ºÍ∏∞Ï†ÅÏúºÎ°ú Ïû¨ÏÑ†ÌÉù)
# ====================================================================
pred_exp = np.full(len(y_test), np.nan, dtype=float)
hist = y_train.copy()
current_order = None

for i in range(len(y_test)):
    if (i % RESELECT_STRIDE == 0) or (current_order is None):
        current_order = select_arima_order(hist, p_max=P_MAX, d_set=D_SET, q_max=Q_MAX)
    fit = fit_arima_safe(hist, current_order, maxiter=300)
    if fit is not None:
        pred_exp[i] = float(fit.forecast(steps=1)[0])
    else:
        pred_exp[i] = float(np.mean(hist[-min(10, len(hist)):] ))
    hist = np.append(hist, y_test[i])

rmse_d = rmse(y_test, pred_exp)
mae_d  = mae(y_test, pred_exp)
r2_d   = r2(y_test, pred_exp)
print(f"[Expanding window] RMSE={rmse_d:.4f} | MAE={mae_d:.4f} | R2={r2_d:.4f}")

plt.figure()
plt.plot(dates_test, y_test, label="Actual (Test)", color="black", linewidth=1.4)
plt.plot(dates_test, pred_exp, label=f"Expanding ARIMA (stride={RESELECT_STRIDE})", linestyle="--")
plt.title(f"{symbol} ‚Äî Expanding-window ARIMA (level)")
plt.xlabel("Date"); plt.ylabel("Price"); plt.xticks(rotation=45); plt.legend(); plt.tight_layout(); plt.show()

print("\n[D] Expanding window ÏÑ§Î™Ö")
print("- ÏãúÍ∞Ñ tÏóêÏÑú 1~t-1ÍπåÏßÄ ÎàÑÏ†Å Îç∞Ïù¥ÌÑ∞Î°ú Îã§Ïãú ÌïôÏäµÌïòÏó¨ 1-step ÏòàÏ∏°.")
print("- Îç∞Ïù¥ÌÑ∞ Ï∂ïÏ†Å ‚Üí ÏïàÏ†ïÏÑ±‚Üë(Î∂ÑÏÇ∞‚Üì), Î≥ÄÌôî Ï†ÅÏùëÏùÄ ÎäêÎ¶¥ Ïàò ÏûàÍ≥† Í≥ÑÏÇ∞Îüâ‚Üë.\n")

# ==================== ÏöîÏïΩ & Ïò§Î≤ÑÎ†àÏù¥ ====================
summary = pd.DataFrame({
    "Method": [
        f"Fixed-origin ARIMA{best_fix}",
        f"Rolling ARIMA (W={W}, stride={RESELECT_STRIDE})",
        f"Expanding ARIMA (stride={RESELECT_STRIDE})"
    ],
    "RMSE": [rmse_b, rmse_c, rmse_d],
    "MAE":  [mae_b,  mae_c,  mae_d],
    "R2":   [r2_b,   r2_c,   r2_d]
})
print("\n=== Summary on Test (level) ===")
display(summary.sort_values("RMSE"))

plt.figure()
plt.plot(dates_test, y_test, label="Actual (Test)", color="black", linewidth=1.6)
plt.plot(dates_test, y_pred_static, label="Fixed-origin", linestyle="--")
plt.plot(dates_test, pred_roll,     label="Rolling",     linestyle=":")
plt.plot(dates_test, pred_exp,      label="Expanding",   linestyle="-.")
plt.title(f"{symbol} ‚Äî Price overlay of 3 evaluation strategies (ARIMA on level)")
plt.xlabel("Date"); plt.ylabel("Price"); plt.xticks(rotation=45); plt.legend(); plt.tight_layout(); plt.show()
```
