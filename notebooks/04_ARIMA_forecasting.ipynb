# # ‚úÖ ARIMA(p, d, q) Î™®Îç∏ÏùÑ ÌôúÏö©Ìïú ÌÖåÏä¨Îùº Ï£ºÍ∞Ä ÏòàÏ∏°

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, r2_score

# üìå 1Ô∏è‚É£ ÌÖåÏä¨Îùº(TSLA) Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥Î°úÎìú (ÏµúÍ∑º 2ÎÖÑÏπò)
df = yf.download("TSLA", start="2022-01-01", end="2024-01-01")
df = df.reset_index()

# üìå 2Ô∏è‚É£ ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï† (80% ÌïôÏäµ, 20% ÌÖåÏä§Ìä∏)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size].copy(), df.iloc[train_size:].copy()

# Îç∞Ïù¥ÌÑ∞ ÌôïÏù∏ (ÎîîÎ≤ÑÍπÖÏö©)
print(f"Train DataFrame: {train.shape}")
print(f"Test DataFrame: {test.shape}")
print(train.head())

# üìå 3Ô∏è‚É£ ÏµúÏ†ÅÏùò (p, d, q) Í∞í Ï∞æÍ∏∞
auto_model = auto_arima(train["Close"], seasonal=False, stepwise=True, trace=True)
best_p, best_d, best_q = auto_model.order
print(f"‚úÖ ÏµúÏ†ÅÏùò (p, d, q) Í∞í: (p={best_p}, d={best_d}, q={best_q})")

# üìå 4Ô∏è‚É£ ARIMA Î™®Îç∏ ÌïôÏäµ
model = ARIMA(train["Close"], order=(best_p, best_d, best_q))
model_fit = model.fit()

# üìå 5Ô∏è‚É£ ÏòàÏ∏° ÏàòÌñâ
train_pred = model_fit.predict(start=best_p, end=len(train) - 1, typ="levels")
test_pred = model_fit.forecast(steps=len(test))  # Î°§ÎßÅ ÏòàÏ∏°Ïù¥ ÏïÑÎãàÎùº, Ï†ÑÏ≤¥ ÏòàÏ∏°

# üìå 6Ô∏è‚É£ ÏÑ±Îä• ÌèâÍ∞Ä
train_actual = train["Close"][best_p:]
test_actual = test["Close"]

train_mse = mean_squared_error(train_actual, train_pred)
test_mse = mean_squared_error(test_actual, test_pred)
train_r2 = r2_score(train_actual, train_pred)
test_r2 = r2_score(test_actual, test_pred)

print(f"üìå Training MSE: {train_mse:.4f}, R¬≤: {train_r2:.4f}")
print(f"üìå Test MSE: {test_mse:.4f}, R¬≤: {test_r2:.4f}")

# üìå 7Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏°)
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Actual", color="black")
plt.plot(test["Date"], test_pred, label="Predicted (ARIMA Model)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA Model")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå 8Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í)
plt.figure(figsize=(12, 5))
plt.plot(train["Date"][best_p:], train_actual, label="Actual (Training)", color="black")
plt.plot(train["Date"][best_p:], train_pred, label="Predicted (Training)", color="blue", linestyle="--")
plt.title("Tesla Stock Price Prediction using ARIMA - Training Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå 9Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í)
plt.figure(figsize=(12, 5))
plt.plot(test["Date"], test_actual, label="Actual (Test)", color="black")
plt.plot(test["Date"], test_pred, label="Predicted (Test)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA - Test Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# ‚úÖ ÏµúÏ†Å (p, d, q) ÌÉêÏÉâ + ARIMA Î™®Îç∏ ÌïôÏäµ + Î°§ÎßÅ ÏòàÏ∏° ÏΩîÎìú

‚úÖ Í∞úÏÑ†Îêú Í∏∞Îä•
ÏµúÏ†ÅÏùò (p, d, q) Í∞í ÏûêÎèô ÌÉêÏÉâ
pmdarima.auto_arima()Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÏµúÏ†ÅÏùò ÌååÎùºÎØ∏ÌÑ∞Î•º ÏûêÎèôÏúºÎ°ú Ï∞æÏùå.
ARIMA Î™®Îç∏ ÌïôÏäµ Î∞è Ï†ÑÏ≤¥ ÏòàÏ∏°
model_fit.predict()Î°ú ÌõàÎ†® Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° ÏàòÌñâ.
Î°§ÎßÅ ÏòàÏ∏° (One-step rolling forecast)
ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î•º Ìïú Ïä§ÌÖùÏî© ÏòàÏ∏°ÌïòÎäî Î∞©Ïãù Ï†ÅÏö©.
ÏÑ±Îä• ÌèâÍ∞Ä Î∞è ÏãúÍ∞ÅÌôî
MSE(Mean Squared Error) Î∞è R¬≤(ÏÑ§Î™ÖÎ†•) Í≥ÑÏÇ∞
ÌõàÎ†® Îç∞Ïù¥ÌÑ∞, ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏòàÏ∏° Í≤∞Í≥º ÏãúÍ∞ÅÌôî.

```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from pmdarima import auto_arima
from sklearn.metrics import mean_squared_error, r2_score

# üìå 1Ô∏è‚É£ ÌÖåÏä¨Îùº(TSLA) Ï£ºÍ∞Ä Îç∞Ïù¥ÌÑ∞ Îã§Ïö¥Î°úÎìú (ÏµúÍ∑º 2ÎÖÑÏπò)
df = yf.download("TSLA", start="2022-01-01", end="2024-01-01")

# ‚úÖ Îã§Ï§ë Ïù∏Îç±Ïä§ Ï†úÍ±∞ (ÌïÑÏàò)
df.columns = df.columns.droplevel(1)  # MultiIndex Ï†úÍ±∞
df = df.reset_index()  # Ïù∏Îç±Ïä§ Ï¥àÍ∏∞Ìôî

# üìå ‚úÖ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ïª¨Îüº ÌôïÏù∏ (ÎîîÎ≤ÑÍπÖ)
print("üîç DataFrame Columns:", df.columns)
print(df.head())

# üìå 2Ô∏è‚É£ ÌïôÏäµ Î∞è ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï† (80% ÌïôÏäµ, 20% ÌÖåÏä§Ìä∏)
train_size = int(len(df) * 0.8)
train, test = df.iloc[:train_size].copy(), df.iloc[train_size:].copy()

# üìå 3Ô∏è‚É£ ÏµúÏ†ÅÏùò (p, d, q) Í∞í Ï∞æÍ∏∞
auto_model = auto_arima(train["Close"], seasonal=False, stepwise=True, trace=True)
best_p, best_d, best_q = auto_model.order
print(f"‚úÖ ÏµúÏ†ÅÏùò (p, d, q) Í∞í: (p={best_p}, d={best_d}, q={best_q})")

# üìå 4Ô∏è‚É£ ARIMA Î™®Îç∏ ÌïôÏäµ (Ï†ÑÏ≤¥ ÏòàÏ∏°Ïö©)
model = ARIMA(train["Close"].astype(float), order=(best_p, best_d, best_q))  # ‚úÖ float Î≥ÄÌôò
model_fit = model.fit()

# üìå 5Ô∏è‚É£ ÏòàÏ∏° ÏàòÌñâ
train_pred = model_fit.predict(start=best_p, end=len(train) - 1, typ="levels")
test_pred = model_fit.forecast(steps=len(test))

# üìå 6Ô∏è‚É£ ÏÑ±Îä• ÌèâÍ∞Ä
train_actual = train["Close"].iloc[best_p:]  # Ïã§Ï†úÍ∞í
test_actual = test["Close"]

train_mse = mean_squared_error(train_actual, train_pred)
test_mse = mean_squared_error(test_actual, test_pred)
train_r2 = r2_score(train_actual, train_pred)
test_r2 = r2_score(test_actual, test_pred)

print(f"üìå Training MSE: {train_mse:.4f}, R¬≤: {train_r2:.4f}")
print(f"üìå Test MSE: {test_mse:.4f}, R¬≤: {test_r2:.4f}")

# üìå 7Ô∏è‚É£ Î°§ÎßÅ ÏòàÏ∏° (One-step rolling forecast)
rolling_predictions = []
history = list(train["Close"].dropna().astype(float))  # ‚úÖ float Î≥ÄÌôò Î∞è NaN Ï†úÍ±∞

# üìå ‚úÖ history Î¶¨Ïä§Ìä∏ ÌôïÏù∏ (ÎîîÎ≤ÑÍπÖ)
print("üîç Initial History Values:", history[:10])

for t in range(len(test)):
    history_array = np.array(history, dtype=float)  # ‚úÖ NumPy Î∞∞Ïó¥ Î≥ÄÌôò
    model = ARIMA(history_array, order=(best_p, best_d, best_q))
    model_fit = model.fit()
    pred = model_fit.forecast(steps=1)[0]  # Ìïú Ïä§ÌÖù ÏòàÏ∏°
    rolling_predictions.append(pred)

    # ‚úÖ `float()` Î≥ÄÌôò ÌõÑ Ï∂îÍ∞Ä (ÎîîÎ≤ÑÍπÖ)
    try:
        history.append(float(test["Close"].iloc[t]))  # float Î≥ÄÌôòÌïòÏó¨ Ï∂îÍ∞Ä
    except ValueError:
        print(f"‚ùå Î≥ÄÌôò Ïã§Ìå®: test['Close'].iloc[{t}] = {test['Close'].iloc[t]}")  # ÏóêÎü¨ Î©îÏãúÏßÄ Ï∂úÎ†•

# üìå 8Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏°)
plt.figure(figsize=(12, 5))
plt.plot(df["Date"], df["Close"], label="Actual", color="black")
plt.plot(test["Date"], rolling_predictions, label="Predicted (ARIMA Model)", linestyle="--", color="red")
plt.title("Tesla Stock Price Prediction using ARIMA Model (Full Forecast)")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå 9Ô∏è‚É£ Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌõàÎ†® Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í)
plt.figure(figsize=(12, 5))
plt.plot(train["Date"].iloc[best_p:], train_actual, label="Actual (Training)", color="black")
plt.plot(train["Date"].iloc[best_p:], train_pred, label="Predicted (Training)", color="blue", linestyle="--")
plt.title("Tesla Stock Price Prediction using ARIMA - Training Phase")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# üìå üîü Í≤∞Í≥º ÏãúÍ∞ÅÌôî (ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞ ÏòàÏ∏° vs Ïã§Ï†úÍ∞í - Î°§ÎßÅ ÏòàÏ∏°)
plt.figure(figsize=(12, 5))
plt.plot(test["Date"], test_actual, label="Actual (Test)", color="black")
plt.plot(test["Date"], rolling_predictions, label="Rolling Predicted (Test)", linestyle="--", color="green")
plt.title("Tesla Stock Price Prediction using ARIMA - Rolling Forecast")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
